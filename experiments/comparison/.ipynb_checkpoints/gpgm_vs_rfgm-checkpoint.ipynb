# ***GPGM vs RFGM Comparison***
---

```julia
using ArgParse
using YAML

using Random
using LinearAlgebra
using Distributions

using DifferentialEquations
using KernelFunctions
using AbstractMCMC
using MCMCChains
using MCMCDiagnosticTools
using StatsBase
using CSV
using DataFrames
using Plots

using EventODEModeling
```

### Load the configuration

```julia
# Execute argparse only when running the script directly
if abspath(PROGRAM_FILE) == @__FILE__
    s = ArgParseSettings()
    @add_arg_table s begin
        "--ode"
            help = "ODE model name"
            arg_type = String
    end
    # Parse the command line arguments
    parsed_args = parse_args(s)
    ode_name = get(parsed_args, "ode", "PredatorPrey")
    config = YAML.load_file(joinpath(dirname(@__FILE__), "..", "gpgm", "config.yaml"))[ode_name]
else
    ode_name = "PredatorPrey"  # Edit here to change ODE when to use notebook
    config = YAML.load_file("../gpgm/config.yaml")[ode_name]
end

true_params       = config["true_parameters"]
u0                = config["initial_conditions"]
tspan             = (config["tspan"][1], config["tspan"][2])
obs_time          = config["obs_time"] .|> float
noise_std         = config["noise_std"]
mcmc_iterations   = config["mcmc_iterations"]
burn_in           = config["burn_in"]
state_noise_std   = config["state_noise_std"]
kernel_type       = config["gp_kernel"]

# Select the ODE
ode_func =
    ode_name == "PredatorPrey" ? EventODEModeling.lotkavolterrapredatorprey! :
    ode_name == "Competition" ? EventODEModeling.lotkavolterracompetition! :
    ode_name == "SIR" ? EventODEModeling.sir! :
    ode_name == "FitzHughNagumo" ? EventODEModeling.fitzhughnagumo! :
    ode_name == "Lorenz96" ? EventODEModeling.lorenz96! :
    ode_name == "SignalTransduction" ? EventODEModeling.signaltransductioncascade! :
    ode_name == "PIF" ? EventODEModeling.pif4and5! : nothing
```

## ***1. ODE Simulation and Data Generation***

```julia
# Set the true parameters and initial conditions
prob = ODEProblem(ode_func, u0, tspan, true_params)
sol = solve(prob, Tsit5(), saveat=0.001)
true_states = Array(solve(prob, Tsit5(), saveat=obs_time))
```

Generate observations by adding noise to the true states

```julia
y_obs = true_states .+ noise_std * randn(size(true_states))
```

plot the true ODE trajectory and the noisy observations

```julia
plot(sol.t, hcat(sol.u...)')
scatter!(obs_time, y_obs', labels=false, title=ode_name, size=(500,200))
```

## Kernel setup

```julia
if kernel_type == "RBF"
    α² = 1.0
    ℓ = 1.0
    k = α² * KernelFunctions.with_lengthscale(KernelFunctions.RBFKernel(), ℓ)
elseif kernel_type == "Matern52"
    α² = 1.0
    ℓ = 1.0
    k = α² * KernelFunctions.with_lengthscale(KernelFunctions.Matern52Kernel(), ℓ)
elseif kernel_type == "Sigmoid"
    α² = 1.0
    a = 0.0
    b = 1.0
    k = α² * SigmoidKernel(b, a)
else
    error("Invalid kernel type: $kernel_type")
end
```

## GPGM setup

```julia
gm_gpgm = GPGM(obs_time, y_obs, prob, ode_name; k=k,
               state_noise_std=state_noise_std, obs_noise_std=noise_std)

priors = [Normal(0., 1.) for _ in 1:length(prob.p)]
transforms = [log for _ in 1:length(prob.p)]
set_priortransform_on_θ!(gm_gpgm, priors, transforms)

optimize_ϕ_and_σ!(gm_gpgm)
optimize_u!(gm_gpgm)
```

```julia
block_X  = HMCBlock(gm_gpgm, [:X]; n_leapfrog=10, step_size=0.05, metric=:diag)
block_Xθ = HMCBlock(gm_gpgm, [:X, :θ]; n_leapfrog=10, step_size=0.01, metric=:diag)
block_θ  = HMCBlock(gm_gpgm, [:θ]; n_leapfrog=10, step_size=0.05, metric=:diag)
bs_gpgm = BlockedSampler([[block_X], [block_Xθ], [block_θ]], [0.4, 0.4, 0.2])

init_gpgm = pack_param_dict(gm_gpgm)

gpgm_runtime = @elapsed begin
    chain_gpgm = AbstractMCMC.sample(gm_gpgm, bs_gpgm, mcmc_iterations;
                                     init_params=init_gpgm, num_burnin=burn_in)
end

θ_samples_gpgm = get_θ(gm_gpgm, chain_gpgm)
```

```julia
gm_gpgm.gp[1].σ
```

```julia
gm_gpgm.gp[1].k
```

```julia
# # Effective Sample Size
θ_chain_gpgm = Chains(θ_samples_gpgm);
MCMCDiagnosticTools.ess(θ_chain_gpgm)
```

## RFGM setup

```julia
gm_rfgm = RFGM(obs_time, y_obs, prob, ode_name; k=k,
               state_noise_std=state_noise_std, obs_noise_std=noise_std)

set_priortransform_on_θ!(gm_rfgm, priors, transforms)

optimize_ϕ_and_σ!(gm_rfgm)
optimize_u!(gm_rfgm)

block_W  = HMCBlock(gm_rfgm, [:W]; n_leapfrog=10, step_size=0.05, metric=:diag)
block_Wθ = HMCBlock(gm_rfgm, [:W, :θ]; n_leapfrog=10, step_size=0.01, metric=:diag)
block_θ  = HMCBlock(gm_rfgm, [:θ]; n_leapfrog=10, step_size=0.05, metric=:diag)
bs_rfgm = BlockedSampler([[block_W], [block_Wθ], [block_θ]], [0.4, 0.4, 0.2])

init_rfgm = pack_param_dict(gm_rfgm)

rfgm_runtime = @elapsed begin
    chain_rfgm = AbstractMCMC.sample(gm_rfgm, bs_rfgm, mcmc_iterations;
                                     init_params=init_rfgm, num_burnin=burn_in)
end

θ_samples_rfgm = hcat([d[:θ] for d in chain_rfgm]...)'
ess_rfgm = minimum(effective_sample_size.(eachcol(θ_samples_rfgm)))
```

## Posterior means and RMSE

```julia
θ_mean_gpgm = mean(θ_samples_gpgm, dims=1)[:]
θ_mean_rfgm = mean(θ_samples_rfgm, dims=1)[:]

rmse_gpgm = sqrt(mean((θ_mean_gpgm .- true_params).^2))
rmse_rfgm = sqrt(mean((θ_mean_rfgm .- true_params).^2))
```

## Save results

```julia
result_df = DataFrame(
    method = ["GPGM", "RFGM"],
    runtime = [gpgm_runtime, rfgm_runtime],
    ess = [ess_gpgm, ess_rfgm],
    rmse = [rmse_gpgm, rmse_rfgm],
)
CSV.write(joinpath(@__DIR__, "results.csv"), result_df)
```

## Plots

```julia
plot(θ_samples_gpgm[:,1], label="GPGM θ₁", xlabel="Iteration", ylabel="θ₁")
plot!(θ_samples_rfgm[:,1], label="RFGM θ₁")
savefig(joinpath(@__DIR__, "trace_theta1.png"))

plot([rmse_gpgm, rmse_rfgm], seriestype=:bar, label=["GPGM" "RFGM"],
     xlabel="Method", ylabel="RMSE", title="θ RMSE")
savefig(joinpath(@__DIR__, "rmse.png"))

result_df
```
